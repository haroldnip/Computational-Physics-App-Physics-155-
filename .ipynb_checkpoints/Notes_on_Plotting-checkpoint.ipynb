This notebook presents fundamentals of plotting using matplotlib.pyplot library. 

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Let's make some synthetic data
num_pts = 1000

dummy_t = np.linspace(0, 1, num_pts)
dummy_x = np.random.random(num_pts)
dummy_y = np.random.randn(num_pts)
```

```python
##Matplotlib's OOP-based plotting. MOre flexible; recommended by matplotlib
fig, ax = plt.subplots()

ax.set(
    xlabel="Describe x-axis",
    ylabel="Descrie y-axis",
    title="Give me a title",
)

fig.tight_layout()
```

```python
## The subplots generator allows you to create multiple plots in a single image. In this example,
## the ax_set returned is a 2x3 array of axes. 

fig, ax_set = plt.subplots(2,3)

print(f"The shape of ax_set is: {ax_set.shape}")

for row in ax_set:
    for ax in row:
        ax.text(0.5, 0.5, r"$e^{-\beta H \{\vec{\sigma}\}}$", color='blue')

## Alternative: iterate over a flattened version
for idx, ax in enumerate(ax_set.flatten()):
    ax.text(0.25, 0.25, f"Enum index:{idx}", color = "red")

fig.tight_layout()
```

```python
fig, ax = plt.subplots()

freq = 4
y_value = np.sin(2*np.pi*freq*dummy_t)
y_value_shift = np.sin(2*np.pi*(freq*dummy_t + 0.25))

ax.plot(dummy_t, y_value)
ax.plot(dummy_t, y_value_shift, color='k', linestyle='-.', linewidth = 0.8)

ax.set(
    xlabel=r"Time $t$",
    ylabel=r"Amplitude $f(t)$",
    xlim=(0.0, 1.0),
    ylim=(-1.25, 1.25)
)

ax.grid()

fig.tight_layout()
```

(Right) Marker size is now variableeach point's size is proportional to its corresponding value in dummy_t. That means points at higher dummy_t values appear larger.

```python
fig, ax = plt.subplots(1, 2, figsize=(8,4))

ax[0].scatter(dummy_t, dummy_y, s=1.0)
ax[1].scatter(dummy_t, dummy_y, s=10*dummy_t, c="black")

fig.tight_layout()
```

```python
fig, ax = plt.subplots(1, 2, figsize=(9, 3))

ax[0].hist(dummy_y, bins = np.linspace(-5, 5, 100))

ax[1].hist(dummy_y, bins = np.linspace(-5, 5, 50), histtype='step', color='black', density=True)
ax[1].hist(dummy_y, bins = np.linspace(-5, 5, 200), histtype='bar', color='red', density=True)

fig.tight_layout()
```
